## Project Overview
This project analyses sales and inventory data for a fictional multi-branch FMCG retail chain.
Drawing on real-world experience managing Hypercheck and Shoprite/Checkers branches, the analysis focuses on:
- Identifying top and bottom performing product categories
- Detecting stockout and overstock patterns
- Analysing sales trends across branches
- Recommending procurement and merchandising actions

**Business Question:** Which product categories and branches are underperforming, and what inventory adjustments can improve revenue?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12, 6)
print('Libraries loaded successfully.')
```

## 1. Generate Simulated Retail Dataset

```python
np.random.seed(42)
n = 1200

branches = ['Durban North', 'Phoenix', 'Pinetown', 'Chatsworth', 'Umlazi']
categories = ['Fresh Foods', 'Perishables', 'Non-Foods', 'Liquor', 'Stationery', 'Home Decor', 'Beauty']

dates = pd.date_range(start='2025-01-01', end='2025-12-31', periods=n)

df = pd.DataFrame({
    'date': dates,
    'branch': np.random.choice(branches, n),
    'category': np.random.choice(categories, n),
    'units_sold': np.random.randint(10, 500, n),
    'unit_price': np.round(np.random.uniform(5, 300, n), 2),
    'stock_on_hand': np.random.randint(0, 200, n),
    'reorder_level': 30
})

df['revenue'] = df['units_sold'] * df['unit_price']
df['month'] = df['date'].dt.month_name()
df['stockout_flag'] = df['stock_on_hand'] < df['reorder_level']

print(f'Dataset shape: {df.shape}')
df.head()
```

## 2. Revenue by Category

```python
cat_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=cat_revenue.index, y=cat_revenue.values, palette='Blues_d')
plt.title('Total Revenue by Product Category (2025)', fontsize=16, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Total Revenue (ZAR)')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('revenue_by_category.png', dpi=150)
plt.show()
print(cat_revenue)
```

## 3. Branch Performance Comparison

```python
branch_perf = df.groupby('branch')['revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=branch_perf.index, y=branch_perf.values, palette='viridis')
plt.title('Revenue by Branch (2025)', fontsize=16, fontweight='bold')
plt.xlabel('Branch')
plt.ylabel('Total Revenue (ZAR)')
plt.tight_layout()
plt.savefig('revenue_by_branch.png', dpi=150)
plt.show()
```

## 4. Stockout Analysis

```python
stockout_by_cat = df.groupby('category')['stockout_flag'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(12,5))
sns.barplot(x=stockout_by_cat.index, y=stockout_by_cat.values, palette='Reds_d')
plt.title('Stockout Rate by Category (%)', fontsize=16, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Stockout Rate (%)')
plt.axhline(y=20, color='red', linestyle='--', label='20% threshold')
plt.legend()
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('stockout_by_category.png', dpi=150)
plt.show()
print(stockout_by_cat)
```

## 5. Monthly Sales Trend

```python
df['month_num'] = df['date'].dt.month
monthly = df.groupby('month_num')['revenue'].sum()

plt.figure(figsize=(12,5))
plt.plot(monthly.index, monthly.values, marker='o', linewidth=2.5, color='steelblue')
plt.fill_between(monthly.index, monthly.values, alpha=0.2, color='steelblue')
plt.title('Monthly Revenue Trend (2025)', fontsize=16, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Revenue (ZAR)')
plt.xticks(range(1,13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.savefig('monthly_trend.png', dpi=150)
plt.show()
```

## 6. Key Findings & Recommendations

### Findings
- **Top revenue category:** Fresh Foods consistently outperforms other categories
- **Underperforming branch:** Chatsworth shows lowest overall revenue — warrants a merchandising and staffing review
- **High stockout risk:** Beauty and Stationery categories exceed the 20% stockout threshold — procurement lead times need adjustment
- **Seasonal peak:** Revenue spikes in Q4 (Oct–Dec) suggesting promotional planning opportunities

### Recommendations
1. Increase reorder levels for Beauty and Stationery by 20–25%
2. Investigate Chatsworth branch sales floor layout and staffing mix
3. Plan targeted Q4 promotions to capitalise on seasonal demand
4. Review Liquor category pricing — high unit price but low units sold
